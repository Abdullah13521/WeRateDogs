# Description

In this project I used python and its data analysis libraries to analyse the tweet data for the WeRateDogs twitter account. I gathered the tweets archive for the WeRateDogs twitter account using Twitter's API. I then used data wrangling skills to assess and clean the data. I have found some interesting insights that I am going to share below.

## Report

```python
# import the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create the data frames
tweets_archive = pd.read_csv('archive_final_clean.csv')
img_predictions = pd.read_csv('img_predictions_final.csv')
```

## Analysis and Visualization

### Compare retweet vs favorite count

```python
tweets_archive.plot('retweet_count','fav_count', kind='scatter')
plt.xlabel('Retweets')
plt.ylabel('Favorites')
plt.title('Favorite count vs Retweets');
```

- There's a positive correlation between retweets and favorite counts but favorite counts seem to be usually higher than retweets.

### Compare the distribution of dog stages

```python
tweets_archive.dog_stage.hist()
plt.xlabel('Dog Stage')
plt.ylabel('Count')
plt.title('Distrubtion of dog stages');
```

- Pupper is the most common stage with 200+ dogs. The mixed stages seem to be uncommon totalling < 25.

### Compare average rating for each dog stage

```python
tweets_archive.groupby('dog_stage')['rating_numerator'].mean().plot()
plt.xlabel('Dog Stage')
plt.ylabel('Average Rating')
plt.title('Average rating for each dog stage');
```

- The puppo stage has the highest rating while pupper has the lowest.

## Compare dog breeds

```python
# filter for the top breeds
top_counts = img_predictions.p1.value_counts().sort_values().tail(10)

top_counts.plot(kind='bar')
plt.title('Count of each dog breed');
```

- The golden retriever seems to be the most common breed with 140+ dogs.

## Compare the count of tweets in each year

```python
tweets_archive.timestamp = pd.to_datetime(tweets_archive.timestamp)
tweets_archive.groupby(tweets_archive.timestamp.dt.year)['index'].count().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of tweets')
plt.title('Number of tweets vs Year');
```

- 2016 seems to be the most active year in the We Rate Dogs account history with 1000+ tweets.

## Compare the count of tweets in each month

```python
tweets_archive.groupby(tweets_archive.timestamp.dt.month)['index'].count().plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Number of tweets')
plt.title('Number of tweets vs Month');
```

- It looks like the account is most active from November to March. They make a big percentage of the total tweets.
